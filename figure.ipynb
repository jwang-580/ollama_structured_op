```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results/evaluation_results_4o_vs_no_section.csv')
plt.figure(figsize=(15, 10))

df_melted = df.melt(id_vars=['field'], 
                    value_vars=['accuracy', 'precision', 'recall', 'f1'],
                    var_name='metric', 
                    value_name='value')

sns.barplot(data=df_melted, 
            x='field', 
            y='value', 
            hue='metric',
            palette='Set2')

plt.xticks(rotation=45, ha='right')
plt.xlabel('Fields')
plt.ylabel('Score')
plt.title('Evaluation Metrics if notes are not sectioned')

plt.tight_layout()

plt.savefig('results/evaluation_metrics_not_sectioned.png', dpi=300, bbox_inches='tight')
plt.show()
```
